```python
MergeD = pd.merge(salesD, CostomerD, on='CustomerKey')
MergeD= pd.merge(MergeD, ProductD, on='ProductKey')
```

```python
fig=plt.figure(figsize=(18,6))
miss_MergeD = pd.DataFrame((MergeD.isnull().sum())*100/MergeD.shape[0]).reset_index()
miss_MergeD["type"] = "MergeD"
ax=sns.pointplot(x="index",y=0,data=miss_MergeD,hue="type")
plt.xticks(rotation=90,fontsize=9)
plt.title("Percentage of Missing values in MergeD")
plt.ylabel("Percentage")
plt.xlabel("Columns")
ax.set_facecolor("k")
fig.set_facecolor("lightgray")
```
